**Visión general creada por IA**

Git is a widely-used, open-source distributed version control system (DVCS) that tracks changes in files and code during software development. It allows multiple people to collaborate on a project, manage different versions of files, and recover earlier versions at any time.


Common **Git Commands** and Workflow: A typical Git workflow involves a few core commands:

| Command |	Purpose|
| :--- | :--- |
|**git init**	| Initializes a new local Git repository in a project folder.|
|**git clone [url]**	| Creates a local copy of a project that already exists on a remote server.|
|**git add [file]**	| Stages a change, adding a snapshot of the file to a staging area, ready for the next commit.|
|**git commit -m -a "[message]"**	| Saves the staged snapshot permanently to the project history with a descriptive message. -m mean  menssage -a mean all|
|**git status**	| Shows the status of changes as untracked, modified, or staged.|
|**git pull**	| Fetches the latest changes from the remote repository and merges them into your local branch.|
|**git push**	| Uploads your local commits to the remote repository (e.g., GitHub, GitLab).|
|**git branch [name]**	| Creates a new branch for isolated development.|
|**git merge [branch]**	| Combines changes from a specified branch into the current active branch.|
|**git checkout [branch]**	| change branch.|

For more information, see the full reference guide to Git commands. https://git-scm.com/docs

Offically documentation: https://git-scm.com/

## Example 1: Start a new repository and publish it to GitHub

First, you will need to create a new repository on GitHub. Do not initialize the repository with a README, .gitignore or License file. This empty repository will await your code.

https://github.com


```python
# create a new directory, and initialize it with git-specific functions
git init my-repo

# change into the `my-repo` directory
cd my-repo

# create the first file in the project. with gedit, nano, touch etc.
touch README.md
#or nano README.md

# git isn't aware of the file, stage it
git add README.md

# take a snapshot of the staging area
git commit -m "add README to initial commit"

# provide the path for the repository you created on github
git remote add origin https://github.com/YOUR-USERNAME/YOUR-REPOSITORY-NAME.git
#or use SSH in your personal computer for only one time
#git@github.com:YOUR-USERNAME/YOUR-REPOSITORY-NAME.git

At this level you have to configurate your git repo and your Personal Access Tokens: PAT

# push changes to github
git push --set-upstream origin main
#or
#git push --set-upstream origin master
#git push origin master
```

### Configurar git

Set your Name and Email (Local): This overrides the global settings for a specific repository only.  

```python
#List all settings: View all the settings Git can find from all levels with git config --list.
git config --list

git config user.name "Project Specific Name"
git config user.email "project.email@example.com"

#Output:
'''user.email=afelipe.rivera@udea.edu.co
user.name=anferivera
core.repositoryformatversion=0
core.filemode=true
core.bare=false
core.logallrefupdates=true
remote.origin.url=git@github.com:anferivera/Fisica_Computacional_1.git
remote.origin.fetch=+refs/heads/*:refs/remotes/origin/*
branch.main.remote=origin
branch.main.merge=refs/heads/main'''
```

```python
#Open config file in editor: Edit configuration files directly in your default text editor
git config --edit

#Output
'''
[core]
        repositoryformatversion = 0
        filemode = true
        bare = false
        logallrefupdates = true
[remote "origin"]
        url = git@github.com:anferivera/Fisica_Computacional_1.git
        fetch = +refs/heads/*:refs/remotes/origin/*
[branch "main"]
        remote = origin
        merge = refs/heads/main
'''
```

### Generate a PAT

A Personal Access Token (PAT) is a secure alternative to using your password for authenticating to GitHub via the command line, with the API, or in other applications. GitHub no longer supports password-based authentication for Git operations and recommends using PATs or other secure methods.

1. Sign in to your GitHub account.
2. Click your profile picture in the top-right corner, then select Settings.
3. In the left sidebar, scroll down and click Developer settings.
4. Under "Personal access tokens" in the left sidebar, click on Tokens (classic).
5. Click Generate new token.
6. Provide a name, set an expiration date (no), and select the scopes or permissions needed for your task. Selec all
7. Click Generate token at the bottom of the page.
8. Immediately copy the token displayed. For security reasons, you will not be able to see it again once you leave the page. **GUARDELO POR FAVOR**.

## Example 2: Contribute to an existing repository in Github

note that is a new banch...

```python
# Download a repository on GitHub to our machine
# Replace `owner/repo` with the owner and name of the repository to clone
#from http
git clone https://github.com/owner/my-repo.git
#from SSH: better in your personal computer. Try later
#git clone git@github.com:owner/repo.git

# change into the `repo` directory
cd my-repo

# create a new branch to store any new changes
git branch my-branch
#Example: git branch andres-1616

# switch to that branch (line of development)
git checkout my-branch

# make changes, for example, edit `file1.md` and `file2.md` using the text editor

# stage the changed files
git add file1.md file2.md

# take a snapshot of the staging area (anything that's been added)
git commit -m "my change"

# push changes to github
git push --set-upstream origin my-branch
```

## Example 3: Contribute to an existing repository

```python
# change into the `repo` directory
cd repo

# update all remote tracking branches, and the currently checked out branch
git pull

#optional
# change into the existing branch called `feature-a`
git checkout feature-a

# make changes, for example, edit `file1.md` using the text editor

# stage the changed file
git add file1.md

# take a snapshot of the staging area
git commit -m "edit file1.md"

# push changes to github
git push
```
