```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = sns.load_dataset("penguins").dropna().values # Cargo los datos, remuevo las filas que tengan valores NaN y los convierto en un arreglo de NumPy
df = pd.DataFrame(data, columns=["especie", "isla", "longitud_pico (mm)", "profundidad_pico (mm)", "longitud_aleta (mm)", "masa_corporal (g)", "genero"]).reset_index(drop=True)
```

```python
Torgensen= df[df['isla']=='Torgersen']
print(f'El máximo de profundidad en el pico es {np.max(Torgensen["profundidad_pico (mm)"])} mm, la profundidad mínima en el pico es {np.min(Torgensen["profundidad_pico (mm)"])} mm')
```

```python
No_Gentoo=df[(df['genero']=='Male') & (df['especie']!='Gentoo')]
```

```python
print(f'La masa corporal media de los pingüinos sin tomar en cuenta la masa de los Gentoo es {np.round(np.mean(No_Gentoo["masa_corporal (g)"]))}')
```

```python
longitud_pico=df['longitud_pico (mm)']
n = len(longitud_pico)
bins = int(1 + np.log2(n))
ndatos_en_i, bordes, grafico = plt.hist(longitud_pico, bins=bins) # Grafico el histograma
centros = 0.5 * (bordes[:-1] + bordes[1:])  #Encuentro el valor medio de cada intervalo
plt.xticks(centros) #Pongo los valores medios en el ege x
plt.title(f'Histograma de longitud de pico, std={np.round(np.std(longitud_pico))} y valor medio={np.round(np.mean(longitud_pico))}')
plt.xlabel('Longitud de pico (mm)')
plt.ylabel('Frecuencia')
plt.axvline(np.mean(longitud_pico), linestyle='--', color='r')
```

std es ~5.5
