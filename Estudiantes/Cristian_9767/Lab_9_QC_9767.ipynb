# Exercise 1: NH₃ molecule

La molécula de amoníaco (NH₃) puede modelarse como un sistema de dos niveles debido al efecto de túnel cuántico del átomo de nitrógeno a través del plano formado por los tres átomos de hidrógeno.

Para esta molécula se tiene una separación energética asociada a la frecuencia

$$
\frac{\Omega}{2\pi}=24.79\ \text{GHz}.
$$

Según el enunciado, la frecuencia de desintonía es

$$
\Delta=\frac{\Omega_R}{2}.
$$

La frecuencia generalizada del sistema está dada por

$$
\Omega=\sqrt{\Omega_R^2+\Delta^2}.
$$

Sustituyendo la condición anterior:

$$
\Omega=\sqrt{\Omega_R^2+\left(\frac{\Omega_R}{2}\right)^2} = \frac{\sqrt5}{2}\Omega_R.
$$

Despejando la frecuencia de Rabi:

$$
\Omega_R=\frac{2}{\sqrt5}\Omega.
$$

A continuación se calcula la frecuencia de Rabi para la molécula de NH₃ y se grafican las probabilidades de ocupación suponiendo que el estado inicial es $|2\rangle$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros del sistema de NH3
f_eff = 24.79e9  # Frecuencia efectiva (Omega / 2*pi) en Hz

# Calcular la frecuencia de Rabi (f_R) y las frecuencias angulares
f_R = (2 / np.sqrt(5)) * f_eff
omega_eff = 2 * np.pi * f_eff
omega_R = 2 * np.pi * f_R

print("=================== RESULTADOS ===================")
print(f"Frecuencia efectiva (Generalizada): {f_eff:.3e} Hz ({f_eff/1e9:.2f} GHz)")
print(f"Frecuencia de Rabi calculada (f_R): {f_R:.3e} Hz ({f_R/1e9:.2f} GHz)")
print("==================================================")

# Simulación de la evolución temporal (estado inicial |2>)
t = np.linspace(0, 5 / f_R, 1000) # Vector de tiempo para 5 ciclos de Rabi

# Calcular las probabilidades de ocupación P2 (estado inicial) y P1 (estado invertido)
delta_over_omega_sq = 1 / 5 # (Delta / Omega)^2 para detuning Delta = Omega_R / 2
P2 = np.cos(omega_eff * t)**2 + (np.sin(omega_eff * t)**2 * delta_over_omega_sq)
P1 = 1 - P2 # Por conservación de la probabilidad

# Gráfica de las Oscilaciones de Rabi
plt.figure(figsize=(9, 5))
plt.plot(t * 1e9, P2, label=r'$P_{|2\rangle}$ (Estado Inicial)', color='#1f77b4', linewidth=2)
plt.plot(t * 1e9, P1, label=r'$P_{|1\rangle}$ (Estado de Inversión)', color='#ff7f0e', linewidth=2)

plt.xlabel('Tiempo (ns)', fontsize=12)
plt.ylabel('Probabilidad de Ocupación', fontsize=12)
plt.title(r'Oscilaciones de Rabi para la molécula de $NH_3$ (Con Detuning $\Delta = \Omega_R/2$)', fontsize=13, pad=15)
plt.xlim(t.min() * 1e9, t.max() * 1e9)
plt.ylim(-0.05, 1.05)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(loc='center right', fontsize=11)
plt.tight_layout()
plt.show()
```

## Resultados

La simulación arroja una frecuencia de Rabi de

$$
f_R = 2.217\times10^{10}\ \text{Hz}
$$

o equivalentemente

$$
f_R \approx 22.17\ \text{GHz}.
$$

La gráfica muestra las oscilaciones de Rabi características de un sistema de dos niveles con desintonía (*detuning*). Como el sistema inicia en el estado $|2\rangle$, la probabilidad de encontrar la molécula en dicho estado comienza en $1$ y disminuye periódicamente únicamente hasta $0.25$, mientras que la probabilidad de ocupación del estado $|1\rangle$ presenta el comportamiento complementario, alcanzando un valor máximo de $0.75$.

Se observa una transferencia periódica e incompleta de población entre ambos estados, debido a que la presencia de la desintonía ($\Delta \neq 0$) penaliza la amplitud máxima de la transición de acuerdo con las ecuaciones de evolución temporal del sistema. A pesar de la atenuación en la amplitud, se conserva en todo momento la normalización de las probabilidades,

$$
P_{|1\rangle}+P_{|2\rangle}=1.
$$

Este comportamiento es consecuencia del acoplamiento cuántico y la interferencia destructiva parcial generada por el *detuning* en los dos estados de inversión de la molécula de NH₃.
